# Visualisasi Data

Library Matplotlib
Yang dapat dilakukan dengan Matplotlib, yakni:

1. Pengolahan dataset (menggunakan library NumPy & Pandas)
2. Membuat grafik dasar menggunakan matplotlib
3. Melakukan modifikasi komponen visualisasi, seperti axis, labels, title, dan legend
4. Menyimpan plot visualisasi yang sudah dibuat



```python
import pandas as pd
dataset = pd.read_csv('gender_submission.csv')
print('Ukuran dataset: %d baris dan %d kolom\n' % dataset.shape)
print('Lima data teratas:')
print(dataset.head())
```

```python
dataset['Survived'].sum()
dataset['Survived'].value_counts()
```

```python
import matplotlib.pyplot as plt
import numpy as np
index = np.arange(len(dataset['Survived'].unique()))
plt.bar(index, dataset['Survived'].sum())
plt.xlabel('Survived', fontsize=12)
plt.ylabel('PassengerId', fontsize=12)
plt.xticks(index, dataset['Survived'].unique(), fontsize=10, rotation=85)
plt.title('Number of who survived in Titanic')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.pie(dataset['Survived'].value_counts(), labels=dataset['Survived'].unique(), startangle=100)
plt.title('Number of Who Survived')
plt.show()
```

```python
dataset.plot(kind='line', figsize=(10,6))
plt.title('Number of Who Survived', size= 16)
plt.ylabel('\nPassengerId', size=14)
plt.xlabel('\nSurvived',size=14)
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(10,6))

dataset1 = dataset.groupby('Survived')
colors = {'Survived':'Red'}

for name, group in dataset1:
  group.plot(ax=ax, kind='scatter', x='Survived', y='PassengerId', s=100)
plt.title('Number of Who Survived', size= 16)
plt.ylabel('\nPassengerId', size=14)
plt.xlabel('\nSurvived',size=14)
plt.show()
```

```python
dataset['Survived'].plot(kind='box', figsize=(8,6))
plt.title('Number of who survived', size=16)
plt.show()
```

```python
dataset['Survived'].plot(kind='hist', figsize=(10,6), bins=50)
plt.title('Number of who survived', size=16)
plt.ylabel('\nSurvived', size=14)
plt.xlabel('\nPassengerId', size=14)
plt.show()
```
